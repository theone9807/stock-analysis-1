```python
df_ohlc = df2['Adj Close'].resample('10D').ohlc() #OHCL open high low close
df_volume = df2['Volume'].resample('10D').sum()

df_ohlc.head()
```

```python
df = web.DataReader('AAPL', 'yahoo', start, end)
print(df.head())
```

```python
#percenntage change f - i / i
i = len(df)
d = {}
for i in range(0, i - 1):
    if df['Adj Close'][i + 1] >= df['Adj Close'][i]:
        df["Daily reaturn"][i]  = (df['Adj Close'][i + 1] - df['Adj Close'][i]) / (df['Adj Close'][i])
        
    elif df['Adj Close'][i + 1] < df['Adj Close'][i]:
        df["Daily reaturn"][i] = (-df['Adj Close'][i + 1] + df['Adj Close'][i]) / (df['Adj Close'][i])
```
